# 8)Unsupervised Learning-Correlation Analysis

__To Find Interdependency of the variables DistanceFrom Home, MonthlyIncome, TotalWorkingYears, YearsAtCompany, YearsWithCurrManager from the Attrition, we did the following Correlation Analysis:__

```python
from scipy.stats import pearsonr as pr
```

```python
d = {'No': 0, 'Yes': 1}
dataset["Attrition"] = dataset['Attrition'].map(d).fillna(dataset['Attrition'])
```

```python
dataset['TotalWorkingYears']=dataset['TotalWorkingYears'].fillna(11.28)
```

```python
def correlation(stats):
    if(stats < 0 and stats >= -0.6):
        return "Moderate Negative correlation"
    elif(stats == -1):
        return "Perfectly Negative correlation"
    elif(stats == 0):
        return "No correlation"
    elif(stats > 0 and stats <= 0.3):
        return "Low Positive correlation"
    elif(stats > 0.3 and stats <= 0.8):
        return "Highly Positive correlation"
    elif(stats == 1):
        return "Perfectly Positive correlation"
```

```python
stats1,p1 = pr(dataset.Attrition,dataset.DistanceFromHome)
print(stats1,p1)
```

```python
stats2,p2 = pr(dataset.Attrition,dataset.MonthlyIncome)
print(stats2,p2)
```

```python
stats3,p3 = pr(dataset.Attrition,dataset.YearsAtCompany)
print(stats3,p3)
```

```python
stats4,p4 = pr(dataset.Attrition,dataset.YearsWithCurrManager)
print(stats4,p4)
```

```python
stats5,p5 = pr(dataset.Attrition,dataset.TotalWorkingYears)
print(stats5,p5)
```
